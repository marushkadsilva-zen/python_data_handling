```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna("Not Available")
```

```python
df.isnull().sum() #identify missing values(all 0s means dataset is clean)
```

```python
important_cols = ['PPID', 'Att8Act', 'EngAct', 'MathsAct', 'EbaccAct']
df_selected = df[important_cols]

df_selected.head()#slicing based on imp cols
```

```python
top_performers = df_selected[
    (df_selected['Att8Act'] >= 4) &
    (df_selected['EngAct'] >= 4) &
    (df_selected['MathsAct'] >= 4)
]

top_performers.head() #top performer
```

```python
top_performers.shape #how many top performers(384 match the criteria)
```

```python
df_selected = df_selected.copy()
```

```python
df_selected['AverageScore'] = df_selected.apply(
    lambda row: (row['Att8Act'] + row['EngAct'] + row['MathsAct']) / 3,
    axis=1
)

df_selected.head() #lambda
```

```python
df_selected['PerformanceLevel'] = df_selected['AverageScore'].map(
    lambda x: "High" if x >= 5 else ("Medium" if x >= 3 else "Low")
)

df_selected.head() #map
```

```python
high_performers = list(
    filter(lambda x: x >= 5, df_selected['AverageScore'])
)

len(high_performers) #filter
```

```python
combined_scores = list(
    zip(df_selected['PPID'], df_selected['AverageScore'])
)

combined_scores[:5] #combined data
```

```python
df_selected['PerformanceLevel'].value_counts().plot(kind='bar')
plt.title("Student Performance Level Distribution")
plt.xlabel("Performance Level")
plt.ylabel("Number of Students")
plt.show()
#count of students in each category
```

```python
plt.hist(df_selected['AverageScore'], bins=10)
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Frequency")
plt.show()
#histogram-shows distribution
```

```python
plt.boxplot(
    [df_selected['EngAct'], df_selected['MathsAct'], df_selected['Att8Act']],
    labels=['English', 'Maths', 'Attainment 8']
)
plt.title("Subject-wise Score Distribution")
plt.ylabel("Scores")
plt.show()
#shows which subject has higher variation
```

```python
plt.scatter(df_selected['Att8Act'], df_selected['AverageScore'])
plt.title("Attainment 8 vs Average Score")
plt.xlabel("Attainment 8 Actual Score")
plt.ylabel("Average Score")
plt.show()
#shows relationship
```

```python
top_10 = df_selected.sort_values(by='AverageScore', ascending=False).head(10)

plt.bar(top_10['PPID'].astype(str), top_10['AverageScore'])
plt.title("Top 10 Students by Average Score")
plt.xlabel("Student PPID")
plt.ylabel("Average Score")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corr_data = df_selected[['Att8Act', 'EngAct', 'MathsAct', 'AverageScore']]
correlation_matrix = corr_data.corr()

correlation_matrix
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap of Student Scores")
plt.show()
```

```python
performance_counts = df_selected['PerformanceLevel'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    performance_counts,
    labels=performance_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Student Performance Level Distribution")
plt.show()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(6, 4))
plt.hist(df_selected['Att8Act'], bins=10)
plt.title("Distribution of Attainment 8 Scores")
plt.xlabel("Attainment 8 Score")
plt.ylabel("Number of Students")
plt.show()
#
```

```python

score_counts = df_selected['Att8Act'].value_counts().sort_index()

plt.figure(figsize=(7, 4))
plt.bar(score_counts.index.astype(str), score_counts.values)
plt.title("Count of Students by Attainment 8 Score")
plt.xlabel("Attainment 8 Score")
plt.ylabel("Number of Students")
plt.show()
#count plot-how many students per score range
```

```python
avg_scores = {
    'English': df_selected['EngAct'].mean(),
    'Maths': df_selected['MathsAct'].mean(),
    'Attainment 8': df_selected['Att8Act'].mean()
}

plt.figure(figsize=(7, 4))
plt.barh(list(avg_scores.keys()), list(avg_scores.values()))
plt.title("Average Scores Across Subjects")
plt.xlabel("Average Score")
plt.show()
#horizontal bar chart
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df_selected['EngAct'], df_selected['Att8Act'])
plt.title("English Score vs Attainment 8 Score")
plt.xlabel("English Score")
plt.ylabel("Attainment 8 Score")
plt.show()
#pairwise scatter -comparision between eng and attainment8
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df_selected['EngAct'], bins=10, alpha=0.7, label='English')
plt.hist(df_selected['MathsAct'], bins=10, alpha=0.7, label='Maths')
plt.title("English vs Maths Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
#density style histograms-compare between english vs maths score distribution
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df_selected['Att8Act'], bins=20, cumulative=True)
plt.title("Cumulative Distribution of Attainment 8 Scores")
plt.xlabel("Attainment 8 Score")
plt.ylabel("Cumulative Count")
plt.show()
#Cumulative Distribution Plot---How scores accumulate across students
```
