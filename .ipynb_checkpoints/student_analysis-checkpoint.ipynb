```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna("Not Available")
```

```python
df.isnull().sum()
```

```python
important_cols = ['PPID', 'Att8Act', 'EngAct', 'MathsAct', 'EbaccAct']
df_selected = df[important_cols]

df_selected.head()
```

```python
top_performers = df_selected[
    (df_selected['Att8Act'] >= 4) &
    (df_selected['EngAct'] >= 4) &
    (df_selected['MathsAct'] >= 4)
]

top_performers.head()
```

```python
top_performers.shape
```

```python
df_selected = df_selected.copy()
```

```python
df_selected['AverageScore'] = df_selected.apply(
    lambda row: (row['Att8Act'] + row['EngAct'] + row['MathsAct']) / 3,
    axis=1
)

df_selected.head()
```

```python
df_selected['PerformanceLevel'] = df_selected['AverageScore'].map(
    lambda x: "High" if x >= 5 else ("Medium" if x >= 3 else "Low")
)

df_selected.head()
```

```python
high_performers = list(
    filter(lambda x: x >= 5, df_selected['AverageScore'])
)

len(high_performers)
```

```python
combined_scores = list(
    zip(df_selected['PPID'], df_selected['AverageScore'])
)

combined_scores[:5]
```

```python
df_selected['PerformanceLevel'].value_counts().plot(kind='bar')
plt.title("Student Performance Level Distribution")
plt.xlabel("Performance Level")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.hist(df_selected['AverageScore'], bins=10)
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Frequency")
plt.show()
```
